### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
# initially set the name of two dataframes based on reading into them the csvs, respectively
Sales_2020 = pd.read_csv('athletic_sales_2020.csv')
Sales_2021 = pd.read_csv('athletic_sales_2021.csv')
# Just to make sure it's actually imported it, I'm going to "lens" the two new dfs
len(Sales_2020)
Sales_2021.shape
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
Sales_2020_2021 = pd.concat([Sales_2020, Sales_2021], axis=0, ignore_index=True)
Sales_2020_2021

# contantenation is more appropriate because all of the columns are the same (as confirmed above)
# join or merge would have been more appropriate if there was a swing item (i.e. a retailer id) that was the same
#  but otherwise the rest of the information was different and I was going to join and merge those two separate files
# with different types of data
# but since the columns are identical, contanenation is more appropriate
```

```python
# Check if any values are null.
# Sales_2020_2021.isnull()
# the above simply reported whether any individual cell is null (it showed the top and bottom 5 as "false"
# will do it as print(df.isnull.sum) for a number
print(Sales_2020_2021.isnull().sum())
```

```python
# Check the data type of each column
# Same as what we did above - 

print(Sales_2020_2021.info())
```

```python
# Convert the "invoice_date" to a datetime datatype
# invoice date is given as an "object" and now well update it to a datetime
# Sales_2020_2021['invoice_date'] = pd.to_datetime(Sales_2020_2021['invoice_date'])
# # What the above did was take the column called "invoice_date" from the df S_2_2 and use panda to convert int a datetime type instead of an object
# # then we'll re-print it to see how it looks
# Sales_2020_2021
# # I got a result that said "To ensure parsing is consisten specify a format
# So redid it as:
Sales_2020_2021['invoice_date'] = pd.to_datetime(Sales_2020_2021['invoice_date'], format= 'mixed')
Sales_2020_2021
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
#  so first I will group things 
Products_Sold_by_Area = Sales_2020_2021.groupby(['region', 'state', 'city'])['units_sold'].sum().reset_index()

# Rename the sum to "Total_Products_Sold".
Products_Sold_by_Area = Products_Sold_by_Area.rename(columns={'units_sold': 'Total_Products_Sold'})



# Show the top 5 results.
Products_Sold_by_Area = Products_Sold_by_Area.sort_values('Total_Products_Sold', ascending=False)
Products_Sold_by_Area = Products_Sold_by_Area.reset_index(drop=True)
Products_Sold_by_Area.head(5)
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
pivot_result = pd.pivot_table(Sales_2020_2021, 
                              values='units_sold', 
                              index=['region', 'state', 'city'], 
                              aggfunc='sum')

# Rename the "units_sold" column to "Total_Products_Sold"
pivot_result = pivot_result.rename(columns={'units_sold': 'Total_Products_Sold'})

pivot_result = pivot_result.reset_index()
# Show the top 5 results.
pivot_result = pivot_result.sort_values('Total_Products_Sold', ascending=False)
print(pivot_result.head(5))
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.

# Rename the "total_sales" column to "Total Sales"


# Show the top 5 results.

Sales_by_Region = Sales_2020_2021.groupby(['region', 'state', 'city'])['total_sales'].sum().reset_index()

# Rename the sum to "Total_Products_Sold".
Sales_by_Region = Sales_by_Region.rename(columns={'total_sales': 'Total_Products_Sold'})



# Show the top 5 results.
Sales_by_Region = Sales_by_Region.sort_values('Total_Products_Sold', ascending=False)
Sales_by_Region = Sales_by_Region.reset_index(drop=True)
Sales_by_Region.head(5)
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.


# Optional: Rename the "total_sales" column to "Total Sales"


# Show the top 5 results.

pivot_result = pd.pivot_table(Sales_2020_2021, 
                              values='total_sales', 
                              index=['region', 'state', 'city'], 
                              aggfunc='sum')

pivot_result = pivot_result.rename(columns={'total_sales': 'Total_Sales'})

pivot_result = pivot_result.reset_index()
# Show the top 5 results.
pivot_result = pivot_result.sort_values('Total_Sales', ascending=False)
print(pivot_result.head(5))
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_athletic_footwear = Sales_2020_2021[Sales_2020_2021["product"].str.contains("Women's Athletic Footwear", case=False, na=False)]
womens_athletic_footwear
# this created a new pseudo-df called "womens_athletic_footwear" that limited itself to rows
#  where the product contained the word's Women's Athletic Footwear
#  I made the case "false" and na responses as 'false' - although we already saw above there are no nulls and the data is pretty clean
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
# Show the top 5 results.
# Will borrow from our "groupby" we used before but this time limited to the Womens_athletic_footwear frame

Sales_by_Retailer_of_Womens_athletic_footwear = womens_athletic_footwear.groupby(['retailer', 'region', 'state', 'city'])['units_sold'].sum().reset_index()

# Rename the sum to "Total_Products_Sold".
Sales_by_Retailer_of_Womens_athletic_footwear = Sales_by_Retailer_of_Womens_athletic_footwear.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'})


# Show the top 5 results.
Sales_by_Retailer_of_Womens_athletic_footwear = Sales_by_Retailer_of_Womens_athletic_footwear.sort_values('Womens_Footwear_Units_Sold', ascending=False)
Sales_by_Retailer_of_Womens_athletic_footwear = Sales_by_Retailer_of_Womens_athletic_footwear.reset_index(drop=True)
Sales_by_Retailer_of_Womens_athletic_footwear.head(5)
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.


# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"


# Show the top 5 results.
# Again we'll borrow but this time for the unique requirements of this task
pivot_result = pd.pivot_table(womens_athletic_footwear, 
                              values='units_sold', 
                              index=['retailer', 'region', 'state', 'city'], 
                              aggfunc='sum')

pivot_result = pivot_result.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'})

pivot_result = pivot_result.reset_index()
# Show the top 5 results.
pivot_result = pivot_result.sort_values('Womens_Footwear_Units_Sold', ascending=False)
print(pivot_result.head(5))
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.


# Optional: Rename the "total_sales" column to "Total Sales"


# Show the table.

pivot_result = pd.pivot_table(womens_athletic_footwear, 
                              values='total_sales', 
                              index='invoice_date', 
                              aggfunc='sum')

pivot_result = pivot_result.rename(columns={'total_sales': 'Total Sales'})

pivot_result = pivot_result.reset_index()
# Show the top 5 results.
pivot_result = pivot_result.sort_values('Total Sales', ascending=False)
print(pivot_result.head(5))
```

```python

print(pivot_result.index.name)
print(pivot_result.columns)
if pivot_result.index.name == 'invoice_date':
    pivot_result.index = pd.to_datetime(pivot_result.index)
else:
    pivot_result = pivot_result.set_index('invoice_date')
    pivot_result.index = pd.to_datetime(pivot_result.index)

daily_sales = pivot_result['Total Sales'].resample('D').sum()

daily_sales_sorted = daily_sales.sort_values(ascending=False)

# Show the top 5 results
print(daily_sales_sorted.head())
```

```python

print(pivot_result.index.name)
print(pivot_result.columns)
if pivot_result.index.name == 'invoice_date':
    pivot_result.index = pd.to_datetime(pivot_result.index)
else:
    pivot_result = pivot_result.set_index('invoice_date')
    pivot_result.index = pd.to_datetime(pivot_result.index)

daily_sales = pivot_result['Total Sales'].resample('W').sum()

daily_sales_sorted = daily_sales.sort_values(ascending=False)

# Show the top 5 results
print(daily_sales_sorted.head())
```
